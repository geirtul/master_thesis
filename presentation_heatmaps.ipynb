```python
def prediction_heatmap_plotter(table, title=None):
    fig, ax = plt.subplots(figsize=(8,5))
    xlabels = ["Linear", "Dense", "CNN", "Pretrained", "Custom"]
    ylabels = ["Single (a)", "Single (b)", "Single (c)", "Double (a)", "Double (b)", "Double (c)"]
    plot = sns.heatmap(table, annot=True, ax=ax, xticklabels=xlabels, yticklabels=ylabels, vmin=0, vmax=1, fmt=".3f")
    plot.set_yticklabels(plot.get_yticklabels(), rotation=0)
    plot.tick_params(axis='both', which='major', labelsize=11, labelbottom = False, bottom=False, top = False, labeltop=True)
    
def classification_heatmap_plotter(table, title=None):
    fig, ax = plt.subplots(figsize=(8,5))
    xlabels = ["Logistic", "Dense", "CNN", "Pretrained", "Custom"]
    ylabels = ["F1-score (a)", "F1-score (b)", "F1-score (c)"]
    plot = sns.heatmap(table, annot=True, ax=ax, xticklabels=xlabels, yticklabels=ylabels, vmin=0, vmax=1, fmt=".3f")
    plot.set_yticklabels(plot.get_yticklabels(), rotation=0)
    plot.tick_params(axis='both', which='major', labelsize=11, labelbottom = False, bottom=False, top = False, labeltop=True)
    
def classification_heatmap_plotter_exp(table, title=None):
    fig, ax = plt.subplots(figsize=(8,5))
    xlabels = ["Single (a)", "Double (a)", "Single (b)", "Double (b)", "Single (c)", "Double (c)"]
    ylabels = ["Logistic", "Dense", "CNN", "Pretrained", "Custom"]
    plot = sns.heatmap(table, annot=True, ax=ax, xticklabels=xlabels, yticklabels=ylabels, vmin=0, vmax=1, fmt=".3f")
    plot.set_yticklabels(plot.get_yticklabels(), rotation=0)
    plot.tick_params(axis='both', which='major', labelsize=11, labelbottom = False, bottom=False, top = False, labeltop=True)
```
